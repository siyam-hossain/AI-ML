```python
# if statement
# if condition is true, then execute the block of code inside the if statement 

age = 23
if age >= 18:
    print("You are an adult.")
```

```python
# else
# The else statement is used to execute a block of code when the condition in the if statement is false.

age = 16
if age >= 18:
    print("You are an adult.")
else:
    print("You are not an adult.")
```

```python
# elif statement
# The elif statement is used to check multiple conditions. It stands for "else if". If the condition in the if statement is false, it checks the condition in the elif statement. If that condition is also false, it checks the next elif statement, and so on. If all conditions are false, it executes the block of code in the else statement.

age = 20

if age < 13:
    print("You are a child.")
elif age < 18:
    print("You are a teenager.")
elif age < 65:
    print("You are an adult.")
else:
    print("You are a senior.")
```

```python
# Nested conditional statements
# You can also nest conditional statements inside each other. This means you can have an if statement inside another if statement.

age = 65
if age >= 18:
    if age < 65:
        print("You are an adult.")
    else:
        print("You are a senior.")
else:
    print("You are not an adult.")
```

```python
# Practical examples
# Determine if a year is a leap year or not
year = int(input("Enter a year: "))

if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print(f"{year} is a leap year.")
        else:
            print(f"{year} is not a leap year.")
    else:
        print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```

```python
# simple calculator
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

operation = input("Enter operation (+, -, *, /): ")

if operation == '+':
    result = num1 + num2
    print(f"The result of {num1} + {num2} is: {result}")
elif operation == '-':
    result = num1 - num2
    print(f"The result of {num1} - {num2} is: {result}")
elif operation == '*':
    result = num1 * num2
    print(f"The result of {num1} * {num2} is: {result}")
elif operation == '/':
    result = num1 / num2
    print(f"The result of {num1} / {num2} is: {result}")
else:
    print("Invalid operation.")
```
