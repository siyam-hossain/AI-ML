**while loop**

```python
# The while loop continues to execute as long as the condition is true. It is used when the number of iterations is not known beforehand.

counter = 0

while counter < 5:
    print(counter, end=' ')
    counter += 1
print(f"\nFinal counter value: {counter}")
```

**loop control statements**

```python
# break
# The break statement exits the loop permanently when a certain condition is met.

for i in range(10):
    if i == 5:
        break
    print(i, end=' ')
print("\nLoop exited.")
```

```python
# continue
# The continue statement skips the current iteration and moves to the next iteration of the loop when a certain condition is met.

for i in range(10):
    if i % 2 == 0:  # Skip even numbers
        continue
    print(i, end=' ')
print("\nLoop completed.")
```

```python
# pass
# The pass statement is a placeholder that does nothing. It is used when a statement is syntactically required but no action is needed.

for i in range(5):
    if i == 2:
        pass  # Placeholder for future code
    print(i, end=' ')
print("\nLoop completed.")
```

```python
# Nested loops
# a loop inside a loop
for i in range(5):
    for j in range(i):
        print("*", end=' ')
    print() 
```

```python
for i in range(3):
    for j in range(2):
        print(f"i: {i}, j: {j}")
    print()  
```

```python
# Examples: calculate the sum of first N natural number using a while and for loop
# while loop
N = 10
sum = 0
counter = 1

while counter <= N:
    sum += counter
    counter += 1
print(f"The sum of first {N} natural numbers is: {sum}")


for i in range(1, N + 1):
    sum += i
print(f"The sum of first {N} natural numbers is: {sum}")
```

<span style="color:#F5BF0F">**for else**</span>

```python
condition1 = False

for i in range(5):
    if condition1:
        break
else:
    # This block executes if the loop completes without encountering a break statement
    print("Loop completed without break.")
```

```python
for i in range(2,2):
    print(i)
```

```python
# Example: prime numbers between 1 and 100

for num in range(1, 101):
    if num > 1:
        for i in range(2, num):
            # num 2, i = 2 : loop will not execute because the range(2, 2) is empty
            # num 3, i = 2 : loop will execute once and check if 3 % 2 == 0 which is false so it will not break and will print 3
            if (num % i == 0):
                break
        else:
            print(num, end=' ')
```
