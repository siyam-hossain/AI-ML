#### **Importing modules in python: modules and packages**

In python, modules and packages help organize and reuse code. Here's a comprehensive guide on how to import them.

```python
import math

print(math.sqrt(16))
```

```python
# from the package_name import any function
from math import sqrt, pi

print(sqrt(16))
print(sqrt(169))
print(sqrt(144))
print(pi)
```

```python
# import package_name as alice 
# alice = custom name of that package_name
import numpy as np
np.array(range(5))
```

```python
# * : mean import all function/methods
from math import *

print(abs(-55))
print(pi)
```

```python
from package.maths import addition

print(addition(4557,668))
```

```python
from package import maths

print(maths.addition(7,3))
```
