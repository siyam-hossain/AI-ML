### **Indentation**

Indentation in python is used to define the structure and hierarchy of the code. Unlike many other programming languages that all the statements within a block must be indented at the same level.

```python
# indentation
# python uses indentation of define blocks of code. Consistent use of spaces (commonly 4 spaces) or a tab is required.

age = int(input("Enter your age: "))

if age >= 18:
    print("You are an adult.")
else:
    print("You are a minor.")
```
