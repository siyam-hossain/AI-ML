```python
# Arithmetic operation

a = 10
b = 5

add_result = a + b              #addtion 
subtract_result = a - b         #subtraction
multiply_result = a * b         #multiplication
divide_result = a / b           #division
floor_divide_result = a // b    #floor division
modulus_result = a % b          #modulus

exponentiation_result = a ** b  #exponentiation

print("Addition: ", add_result)
print("Subtraction: ", subtract_result)
print("Multiplication: ", multiply_result)
print("Division: ", divide_result)
print("Floor Division: ", floor_divide_result)
print("Modulus: ", modulus_result)
print("Exponentiation: ", exponentiation_result)
```

**Comparison Operators**

```python
# Comparison operation
# Equal to (==)
a = 10
b = 10

print("Equal to (a==b): ", a == b)  # True

str1 = "Siyam"
str2 = "siyam"
print("Equal to (str1==str2): ", str1 == str2)  # True
```

```python
# Not equal to (!=)
print("Not equal to (a!=b): ", a != b)  # False
print("Not equal to (str1!=str2): ", str1 != str2)
```

```python
# Greater than (>)
num1 = 15
num2 = 10

print("Greater than (num1>num2): ", num1 > num2)  # True
```

```python
# less than (<)
print("Less than (num1<num2): ", num1 < num2)  # False
```

```python
# Greaer than or equal to (>=)
num1 = 10
num2 = 5

print("Greater than or equal to (num1>=num2): ", num1 >= num2)  # True
```

```python
# Less than or equal to (<=)

a = 45
b = 50

print("Less than or equal to (a<=b): ", a <= b)  # True
```

**Logical Operators**

```python
# AND, OR, NOT

a = True
b = True

and_result = a and b  # True
or_result = a or b    # True
not_result = not a    # False

print("a AND b: ", and_result)
print("a OR b: ", or_result)   
print("NOT a: ", not_result)
```

```python
a = False
b = True

and_result = a and b  # False
or_result = a or b    # True     
not_result = not a    # True

print("a AND b: ", and_result)
print("a OR b: ", or_result)
print("NOT a: ", not_result)
```
