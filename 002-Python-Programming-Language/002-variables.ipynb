**Naming Conventions**
+ Naming conventions
+ variables should be descriptive
+ They must start with a letter or and underscore '_'
+ and contian only letters, numbers and underscores
+ variables are case sensitive

```python
# valid variable names
first_name = "Siyam"
last_name = "Hossain"

# invalid variable names (uncommenting these will cause errors)
# 1name = "Siyam"  # Cannot start with a number
# my-name = "Siyam"  # Cannot contain hyphens
# class = "Math"  # Cannot use reserved keywords
# @name = "Siyam"  # Cannot start with special characters
```

**Variable Types**

python is a dynamically typed



```python
age = 23 #int
height = 5.8 #float
name = "Siyam" #string
is_student = True #boolean

print(type(age)) # Output: <class 'int'>
print(type(height)) # Output: <class 'float'>   
print(type(name)) # Output: <class 'str'>
print(type(is_student)) # Output: <class 'bool'>
```

**Type of checking and Conversion**

```python
print(age)
print(type(age))

# type conversion
age_str = str(age)  # Convert int to string
print(age_str)  # Output: "23"
print(type(age_str))  # Output: <class 'str'>
```

```python
age = '23'
print("age: ", age)  # Output: age:  23
print(type(age))  # Output: <class 'str'>

print("age: ", int(age))  # Output: age:  23
print(type(int(age)))  # Output: <class 'int'>
```

```python
# Dynamically typing 
# Python is allows the type of a variable to change as the program executes
var = 10
print(var, type(var))  

var = "Now I'm a string"
print(var, type(var))

var = 3.14
print(var, type(var))

var = True
print(var, type(var))
```
