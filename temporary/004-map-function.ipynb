#### **The map() function in python**

The map function applies a given function to all items in an input list (or any other iterable) and returns a map object (an iterator). This is particularly useful for transforming data in a list comprehensively.

```python
map(function, iterator)
```

+ remember use function name without bracket

````python
def square(x):
    return x**2

print(square(5))
print(square(10))
print(square(13))
````

````python
numbers = list(range(1,9))

print(map(square, numbers))

# if i want result in the form of list
print(list(map(square, numbers)))
# if i want result in the form of tuple
print(tuple(map(square, numbers)))
# unpack value only
print(*tuple(map(square, numbers)))
````

+ lambda function with map()

````python
numbers = list(range(1,10))

list(map(lambda x: pow(x,2), numbers))
````

+ map multiple iterables

````python
numbers1 = list(range(1,4))
numbers2 = list(range(4,7))

added_numbers = list(map(lambda x,y: x+y, numbers1,numbers2))

print(added_numbers)
````

+ map() to convert a list of strings to integers

````python
str_numbers = ['1','2','3', '4', '5']

# int = int() type converting function
int_numbers = list(map(int,str_numbers))
print(int_numbers)
````

````python
words = ['apple','banana','orange']
upper_words = list(map(str.upper, words))
print(upper_words)
````

````python
words = ['apple','banana','orange']
capitalize_words = list(map(str.capitalize, words))
print(capitalize_words)
````

+ applied of list of dictionary

````python
def get_name(person):
    return person['name']

# list(dictionary)
people = [
    # iterable item 1
    {
        'name' : 'siyam',
        'age': 23
    },
    # iterable item 2
    {
        'name' : 'hossain',
        'age': 21
    }
]

list(map(get_name, people))
````

**Conclusion**

The map() function is a powerful tool for applying transformations to iterable data structures. It can be used with regular functions, lambda functions, and even multiple iterables, providing a versatile approach to data processing in python. By understanding and utilizing map(), you can write more efficient and readable code.
