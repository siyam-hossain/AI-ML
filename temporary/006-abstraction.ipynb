#### **Abstraction**

Abstraction is the concept of hiding the complex implementation details and showing only the necessary features of an object. This helps in reducing programming complexity and effort.

Remember:
+ if an abstract base class contains an abstract method then
+ while child class inherit that class, child class must use that abstract method own way
+ other then it through an error

```python
from abc import ABC, abstractmethod

# abstract base class
class Vehicle(ABC):
    def drive(self):
        print("The vehicle is used for driving")
    
    # hey there is a functionality which is available for you 
    @abstractmethod
    def start_engine(self):
        pass

# child class
class Car(Vehicle):
    def start_engine(self):
        print("car engine started")
    
    def output(self):
        print("hello world")

def operate_vehicle(vehicle_object):
    vehicle_object.start_engine()


car = Car()
car.output()
operate_vehicle(car)    
```

+ what if abstract method is not used in child class

```python
from abc import ABC, abstractmethod

# abstract base class
class Vehicle(ABC):
    def drive(self):
        print("The vehicle is used for driving")
    
    # hey there is a functionality which is available for you 
    @abstractmethod
    def start_engine(self):
        pass

# child class
class Car(Vehicle):

    # without abstract method

    def output(self):
        print("hello world")

car = Car()
car.output()
```
