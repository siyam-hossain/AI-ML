#### **Introduction to tuples**

Explanation:
Tuples are ordered collections of items that are immutable. They are similar to lists, but immutability makes them different.

```python
# creating a tuple

empty_tuple = ()
print(empty_tuple)
print(type(empty_tuple))
```

```python
# converting list into tuple
numbers = tuple(
    [1,2,3,4]
)
numbers
```

```python
# converting tuple into list
list(
    (1,2,3,4)
)
```

```python
mixed_tuple = (1, "Hello world", 3.1416, True)
print(mixed_tuple)
```

**Immutable Nature of Tuples**

Tuples are immutable, meaning their elements cannot be changed once assigned.

```python
# list are mutable mean values can be changed
lst = [1,2,3,4,5]
print(lst)

lst[1] = "Siyam Hossain"
print(lst)
```

```python
numbers
```

```python
numbers[0] = "siyam"
print(type(numbers))
print(numbers)
```

**Tuple Methods**

```python
numbers = (1,2,3,1,1,34,5,67,75,43,43,43,32,7,7);
print(numbers)

print(numbers.count(1))
```

```python
# find the index of 34, it returns the first index if a number appear multiple time
print(numbers.index(34))
```

**Packing and unpacking tuple**

```python
packed_tuple = 1, "hello world", 3.14
print(packed_tuple)
```

```python
# unpacking a tuple
a,b,c = packed_tuple
print(a)
print(b)
print(c)
```

```python
# unpacking with *
numbers = tuple(range(1,7))

first, *middle, last = numbers

print(type(numbers))
print(first)
print(middle)
print(last)
```

```python
print(*numbers)
```

**Nested tuples**

```python
nested_tuple = (
    (1,2,3),
    ("a", "b", "c"),
    (True, False)
)

# access the elements inside a tuple
print(nested_tuple[0])
# output c
print(nested_tuple[1][2])
```

**iterating over nested tuples**

```python
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item, end=" ")
    print()
```

```python
# second column only
for row in nested_tuple:
    # print(row, end=" \n")
    print(row[1], end=" \n")
```

**Conclusion**

Tuples are versatile and useful in many real-world scenarios where an immutable and ordered collection of items is required. They are commonly used in data structures, function arguments and return values, and as dictionary keys, Understanding how to leverage tuples effectively can improve the efficiency and readability of your python code.
