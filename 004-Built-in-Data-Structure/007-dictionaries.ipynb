#### **Introduction to dictionaries**

Dictionaries are unordered collections of items. They store data in key-value pairs. keys must be unique and immutable (e.g., string, numbers, or tuples), while values can be of any type.

**Creating dictionaries**

```python
empty_dict = {}
print(empty_dict)
print(type(empty_dict))
```

```python
empty_dict = dict()
print(empty_dict)
```

```python
student = {
    "name":"siyam hossain",
    "age":23,
    "cgpa": 3.95
}

print(student)
print(type(student))
```

```python
# single key is always used instead of multiple, otherwise new one replace with previous one
student = {
    "name":"siyam hossain",
    "name":24,
    "age":23,
    "cgpa": 3.95
}

print(student)
print(type(student))
```

+ accessing dictionary elements

```python
student = {
    "name":"siyam hossain",
    "age":23,
    "cgpa": 3.95,
    "sex": 'M',
    "email":"siyam.cybersoul@gmail.com"
}
print(student)
```

```python
# accessing dictionary elements
print(student['name'])
print(student['age'])
print(student['cgpa'])
```

```python
# accessing using get() methods
print(student.get('cgpa'))
print(student.get('email'))

# accessing key which is not present
print(student.get("mobile"))

# accessing key which is not present, adding a default value on it
print(student.get("mobile", "Not available"))
```

+ modifying dictionary elements

```python
# dictionary are mutable, so you can add, update or delete elements
print(student)
```

+ update a key value

```python
student["age"] = 24
print(student)
```

+ add a new key

```python
student["address"] = "Bangladesh"
print(student)
```

+ delete a key

```python
print(student)
del student['email']
print(student)
```

+ dictionary methods

```python
# all keys
keys = student.keys()
print(keys)
```

```python
# all values
values = student.values()
print(values)
```

```python
# key value pair
items = student.items()
print(items)
print(type(items))
```

```python
# converted into list
items = list(student.items())
print(items)
print(type(items))
```

**Shallow copy**

```python
# shallow copy
student_copy = student

print(student)
print(student_copy)
```

```python
student["name"] = "Bob"
print(student)
print(student_copy)
```

+ + vital point here

```python
# shallow copy
student_copy_1 = student.copy()
print(student)
print(student_copy_1)
```

```python
student["name"] = "Alice"
print(student)
print(student_copy_1)
```

**Iterating over dictionaries**

+ you can use loops to iterate over dictionaries, keys, values, or items

```python
# iterating over keys
for keys in student.keys():
    print(keys)
```

```python
# iterating over values
for values in student.values():
    print(values)
```

```python
# iterating over key value pairs
for items in student.items():
    print(items)
```

```python
# iterating over key value pairs
for key, value in student.items():
    print(f"{key} : {value}")
```

**Nested Dictionaries**

```python
students = {
    "student1":{
        "name":"siyam",
        "age":23,
    },
    "student2":{
        "name":"bob",
        "age":33,
    }
}
print(students)
```

+ accessing nested dictionaries elements

```python
print(students['student1']['name'])
print(students['student1']['age'])
```

+ iterating over nested dictionary

```python
students.items()
```

```python
for student_id, student_info in students.items():
    # print("-"*50)
    print(f"{student_id}: {student_info}")
    for key, value in student_info.items():
        print(f"{key} : {value}")
    print("-"*50)
```

**Practical example**
+ use a dictionary to count the frequency of elements in list

```python
numbers = [1,2,3,3,3,4,4,4,4]
frequency = {}

for number in numbers:
    if number in frequency:
        # frequency[key] = frequency[key] + value
        frequency[number] += 1
    else:
        # frequency[key] = value
        frequency[number] = 1

print(frequency)
```

```python
# merge 2 dictionaries
dict1={'a':1, 'b':2}
dict2={'b':3, 'c':4}

# using keyword arguments
merged_dict = {**dict1, **dict2}
print(merged_dict)
```
