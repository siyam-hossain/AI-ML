#### **Sets**

Sets are a built-in data type in python used to store collections of unique items. They are unordered, meaning that the elements do not follow a specific order, and that do not allow duplicate elements. Sets are useful for membership tests, eliminating duplicate entries, and performing mathematical set operations like union, intersection, difference, and symmetric difference.

**Creating a set**

```python
my_set = {1,2,4,5,6}
print(my_set)
print(type(my_set))
```

```python
my_empty_set = set()
print(my_empty_set)
print(type(my_empty_set))
```

```python
set_1 = set(
    [1,2,3,4]
)
print(set_1)
print(type(set_1))
```

```python
set_is_unique = {1,2,2,2,3,3,3,3,'siyam','siyam','Siyam'}
print(set_is_unique)
```

**Basic set operations**

+ Adding and removing

```python
my_set
```

```python
my_set.add(7)
print(my_set)

# adding same value: which will ignore
my_set.add(7)
print(my_set)
```

+ remove the elements from a set

```python
print(f"before remove: {my_set}")
my_set.remove(2)
print(f"after remove: {my_set}")
```

```python
# what if we want to remove element which is not present in set
print(f"before remove: {my_set}")
my_set.remove(3)
print(f"after remove: {my_set}")
```

```python
# to avoid the above error
print(f"before remove: {my_set}")
my_set.discard(3)
print(f"after remove: {my_set}")
```

+ pop method

```python
print(f"before pop: {my_set}")

removed_element = my_set.pop()

print(f"pop element: {removed_element}")
print(f"after pop: {my_set}")
```

```python
print(f"before pop: {my_set}")

removed_element = my_set.pop()

print(f"pop element: {removed_element}")
print(f"after pop: {my_set}")
```

+ clear all the elements

```python
print(f"before clear: {my_set}")

my_set.clear()

print(f"after clear: {my_set}")
```

+ set membership test

```python
my_set = set(range(1,7))
print(my_set)
```

```python
print(3 in my_set) # 3 is present in the set or not
print(10 in my_set) # 10 is present in the set or not
```

+ mathematical operation

```python
set1 = set(range(1,7))
set2 = set(range(4,10))

print(set1)
print(set2)
```

```python
# union
union_set = set1.union(set2)
print(union_set)
```

```python
# intersection
intersection_set = set1.intersection(set2)
print(intersection_set)
```

```python
# intersection result update in set1; update method present for all type of set operation
set1.intersection_update(set2)
print(set1)
```

```python
set1 = set(range(1,7))
set2 = set(range(4,10))
```

```python
# difference : common elements are being removed
print(set1.difference(set2))
```

+ symmetric difference

```python
# common elements are removed and uncommon elements are being merged
print(set1.symmetric_difference(set2))
```

+ set methods

```python
set1 = set(range(1,9))
set2 = set(range(3,6))

print(set1)
print(set2)
```

```python
# is subset : mean set1 is parent of set2 or not
print(set1.issubset(set2))
```

```python
# super set
print(set1.issuperset(set2))
```

```python
lst = [1,2,2,3,4,4,5]

set(lst)
```
