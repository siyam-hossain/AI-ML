### **The filter() function in python**

The filter() function constructs an iterator from elements of an iterable for which a function returns true. It is used to filter out items from a list (or any other iterable) based on a condition.


syntax:
```python
filter(function, iterable)
```

+ if function true, only that value will return

````python
def even(num):
    if num%2==0:
        return True
    return False

print(even(12))
print(even(11))
````

````python
lst = list(range(1,13))
print(lst)

list(filter(even, lst))
````

+ when function return false

````python
def even(num):
    return False

print(even(12))
print(even(11))
````

````python
lst = list(range(1,13))
print(lst)

list(filter(even, lst))
````

+ filter with lambda function

````python
numbers = list(range(1,10))

greater_than_five = list(filter(lambda x: x>5, numbers))

print(greater_than_five)
````

+ filter with lambda function and multiple condition

````python
numbers = list(range(1,10))

even_and_greater_than_five = list(filter(lambda x: x>5 and x%2==0, numbers))

print(even_and_greater_than_five)
````

+ apply filter to check if the age is greater than 25 in dictionaries

````python
# list(dictionary)
people = [
    # iterable item 1
    {
        'name' : 'siyam',
        'age': 23
    },
    # iterable item 2
    {
        'name' : 'hossain',
        'age': 21
    },
    # iterable item 3
    {
        'name' : 'alice',
        'age': 27
    },
    # iterable item 4
    {
        'name' : 'bob',
        'age': 49
    }
]

def age_greater_than_25(person):
    return person['age']>25

age_gt25 = list(filter(age_greater_than_25, people))
print(age_gt25)

````

**Conclusion**

The filter() function is a powerful tool for creating iterators that filter items out of an iterable based a function. It is commonly used for data cleaning, filtering objects, and removing unwanted elements from lists. By mastering filter(), you can write more concise and efficient code for processing adn manipulating collections in python.
