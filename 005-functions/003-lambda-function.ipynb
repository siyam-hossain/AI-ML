#### **Lambda Functions in python**

Lambda functions are small anonymous functions defined using the lambda keyword. They can have any number of arguments but only one expression. They are commonly used for short operations or as arguments to higher-order functions.

+ anonymous function : function without name
+ hold only one expression

```python
lambda arguments: expression
```

````python
def addition(a,b):
    return a+b

print(addition(5,887))
````

````python
# lambda function
addition = lambda a,b : a+b

print(type(addition))

# calling the lambda function
print(addition(4587, 6448))
````

````python
def even(num):
    if num%2==0:
        return True
    return False
print(even(24))
print(even(25))
````

````python
even_lambda = lambda num: num%2==0

print(even_lambda(21))
print(even_lambda(22))
````

````python
def addition(x,y,z):
    return x+y+z

print(addition(5,8,9))
````

````python
addition_sum_3 = lambda a,b,c: a+b+c

print(addition_sum_3(5,8,9))
````

````python
print(range(5))
print(*range(5))
print(list(range(5)))
````

````python
def addition_positional_arg(*args):
    print(type(args))
    print(args)
    sum = 0
    for num in args:
        sum += num
    return sum

print(addition_positional_arg(*range(5)))
````

````python
print(addition_positional_arg(*tuple(range(5))))
````

````python
print(addition_positional_arg(*[x for x in range(10)]))
````

**Unpacking**

````python
list1 = [x for x in range(10)]
print(*list1)
print(*[x for x in range(10)])
````

+ map() : applies a function to all items in a list

```python
map(function, iterable)
```

````python
def square(number):
    return number**2

print(square(2))
print(square(5))
````

````python
num_list = list(range(2,20,2))
print(num_list)

list(map(lambda x : x**2, num_list))
````
