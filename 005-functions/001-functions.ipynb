**Introduction to functions**

Definition:

A function is a block of code that performs a specific task. Functions help in organizing code, reusing code, and improving readability.


````python
# why functions?
num = 21
if num%2==0:
    print("the number is even")
else:
    print("the number is odd")
````

**Variable length arguments**

+ positional argument
+ keywords arguments

**Positional arguments**

````python
# positional arguments
def print_numbers(*siyam):
    for number in siyam:
        print(number, end=" ")


print_numbers(1,2,3,4,5,4,6,7,4,2.3,True,"hey there")
````

````python
# positional arguments: best practice
def print_numbers(*args):
    for number in args:
        print(number, end=" ")

        
print_numbers(1,2,3,4,5,4,6,7,4,2.3,True,"hey there")
````

**keyword arguments**

````python
# keywords arguments: all the parameter will be key:value pair format
def print_details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}:{value}")
````

````python
print_details(name="siyam", age="21", country="Bangladesh")
````

````python
def print_details(*args,**kwargs):

    print("positional arguments:",end=" ")
    for val in args:
        print(val, end=" ")
    print()
    

    print("keyword arguments:")
    for key, value in kwargs.items():
        print(f"{key}:{value}")
````

````python
print_details(1,2,3,4,5,"siyam",name="siyam", age="21", country="Bangladesh")
````
