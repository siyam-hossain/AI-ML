#### **Standard library overview**

Python's standard library is a vast collection of modules and packages that come bundled with python, providing a wide range of functionalities out of the box. Here's an overview of some of the most commonly used modules and packages in the python standard library

```python
import array

arr = array.array('i',[1,2,3,4,5])
print(arr)
```

```python
import math
print(math.pi)
```

```python
# random
import random

print(random.randint(1,10))
print(random.choice(['apple','orange','banana','mango']))
```

```python
# file and directory access
import os

# cwd mean : current working directory
print(os.getcwd())
```

```python
os.mkdir('temp-folder')
```

+ high level operations on files and collection of files

```python
import shutil

shutil.copyfile('002-source.txt','002-destination.txt')
```

+ data serialization

```python
import json
data = {
    'name' : 'Siyam',
    'age' : 21
}

json_string = json.dumps(data)
print(json_string)
print(type(json_string))

parsed_data = json.loads(json_string)
print(parsed_data)
print(type(parsed_data))
```

+ csv

```python
import csv

with open('002-example.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['name','age'])
    writer.writerow(['siyam',21])

with open('002-example.csv', mode='r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

+ date time

```python
from datetime import datetime, timedelta

now = datetime.now()
print(f"now: {now}")

yesterday = now-timedelta(days=1)
print(f"yesterday: {yesterday}")
```

+ time

```python
import time

print(time.time())
time.sleep(2)
print(time.time())
```

+ regular expression

```python
import re

pattern = r'\d+'
text = 'there are 123 apples 4568'
match = re.search(pattern, text)

print(match.group())
print(match)
```

**Conclusion**

Python's standard library is extensive and provides tools for almost any task you can think of, from file handling to web services, from data serialization to concurrent execution. Familiarizing yourself with the modules and packages available in the standard library can significantly enhance your ability to write efficient and effective python programs.
