#### **SQL and SQLite**

SQL is a standard language for managing and manipulating relational databases. SQLite is sa self-contained, serverless, and zero-configuration database engine that is widely ued for embedded database systems. in this lesson, we will cover the basics of SQL and SQlite, including creating databases, tables, and performing various SQL operations.

```python
import sqlite3
```

---
+ connect to an sqlite database

```python
connection = sqlite3.connect('001-example.db')
connection
```

---
+ creating cursor

```python
cursor = connection.cursor()
```

---
+ creating a table

```python
cursor.execute('''
Create Table If not Exists employees(
            id integer primary key,
            name text not null,
            age integer, 
            department text
        )
''')
```

+ commit the changes

```python
connection.commit()
```

---

```python
cursor.execute('''
    select * from employees

''')
```

---
+ insert data in db

```python
cursor.execute('''
    insert into employees(name, age, department)
               values('siyam',23,'cse')

''')
cursor.execute('''
    insert into employees(name, age, department)
               values('alice',23,'cse')

''')
cursor.execute('''
    insert into employees(name, age, department)
               values('bob',23,'cse')

''')

connection.commit()
```

---
+ Query the data form the table 

```python
cursor.execute(
    '''
        select * from employees
    '''
)

rows = cursor.fetchall()

# print the queried data

for row in rows:
    print(row)
```

---
+ update the data in the table

```python
cursor.execute(
'''
    update employees
    set age = 21
    where name='siyam'
'''
)
connection.commit()
```

```python
cursor.execute(
    '''
        select * from employees
    '''
)

rows = cursor.fetchall()

# print the queried data

for row in rows:
    print(row)
```

---
+  delete the data from the table

```python
cursor.execute(
'''
    delete from employees 
    where name = 'bob'
'''
)
connection.commit()
```

```python
cursor.execute(
'''
    select * from employees
'''
)

rows = cursor.fetchall()

# print the queried data

for row in rows:
    print(row)
```

**working with sales data**
+ connect to an sqlite database

```python
connection = sqlite3.connect('001-sales-data.db')
cursor = connection.cursor()
```

```python
cursor.execute(
'''
    create table if not exists sales(
        id integer primary key,
        date text not null,
        product text not null,
        sales integer,
        region text
    )
'''
)
```

```python
sales_data = [
    ('2023-01-01', 'P1', 100, 'North'),
    ('2023-01-02', 'P2', 200, 'South'),
    ('2023-01-03', 'P1', 150, 'East'),
    ('2023-01-04', 'P3', 250, 'West'),
    ('2023-01-05', 'P2', 300, 'North'),
]
```

+ bulk insertion

```python
cursor.executemany(
'''
    insert into sales(date, product, sales, region)
    values(?,?,?,?)
''',sales_data
)

connection.commit()
```

```python
cursor.execute(
'''
    select * from sales
'''
)

rows = cursor.fetchall()

for row in rows:
    print(row)
```

---
+ close database connection

```python
connection.close()
```
