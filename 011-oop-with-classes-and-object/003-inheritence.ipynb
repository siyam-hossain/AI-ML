#### **Inheritance in python**

inheritance is a fundament concept in oop that allows a class to inherit attributes and methods from another class. This lesson covers single inheritance and multiple inheritance, demonstrating how to create and use them in python.

````python
# single inheritance
# parent class

class Car:
    def __init__(self, windows, doors, engine_type):
        self.windows = windows
        self.doors = doors
        self.engine_type = engine_type

    def drive(self):
        print(f"the person will dive the {self.engine_type} car")
        
````

````python
car1 = Car(4,4,"Petrol")
print(car1.doors)
print(car1.windows)
print(car1.engine_type)

car1.drive()
````

+ super().__init__(parent attributes) : calling parents class

````python
class Tesla(Car):
    def __init__(self, windows, doors, engine_type, is_self_driving):
        # calling parent init methods over here
        super().__init__(windows, doors, engine_type)

        # current class attributes after inheriting parent class
        self.is_self_driving = is_self_driving
    
    def self_driving(self):
        print(f"Tesla supports self driving: {self.is_self_driving}")
        
````

````python
tesla1 = Tesla(4,4,"ev",True)

tesla1.self_driving()
# calling inherited method
tesla1.drive()
````

+ **Multiple inheritance**
  + When a class inherits from more than one base class.

````python
# base class 1
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print("subclass must implement this method")

# base class 2
class Pet:
    def __init__(self, owner):
        self.owner = owner

# Drive class
class Dog(Animal, Pet):
    def __init__(self, name, owner):
        # instead of writing super(). we use class_name.
        Animal.__init__(self,name)
        Pet.__init__(self, owner)
    
    def speak(self):
        return f"{self.name} say woof"
````

````python
dog = Dog('Tommy','Siyam')
# own method get higher priority, rather than multi inheritance class
# kind of method overriding 
dog.speak()
````

````python
print(f"dog name [from->Animal class]: {dog.name}, owner [from->Pet class]: {dog.owner}")
````
