#### **Custom exception (Raise and Throw an exception)**

```python
class Error(Exception):
    pass

# date of birth exception
class dobException(Error):
    pass
```

```python
year = int(input("Enter your Date of birth: "))
age = 2026 - year

try:
    if age <= 30 and age >= 20:
        print("the age is valid so you can apply for exam")
    else:
        raise dobException
except dobException:
    print("sorry, your age should be grater than 20 or less than 30")
```
