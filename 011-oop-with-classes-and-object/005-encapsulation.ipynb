#### **Encapsulation**

Encapsulation is the concept of wrapping data (variables) and methods (functions) together as a single unit. It restricts direct access to some of the object;s components, which is a means of preventing accidental interference and misuse of the data.


+ Encapsulation with getter and setter methods
  + Three important access variable or access modifier
    + public : use anywhere you want
    + private
    + protected

+ public access modifier
  + variable that can access and modified outside of the class

```python
class Person:
    def __init__(self, name, age):
        self.name = name #public variable 
        self.age = age # public variable 

person = Person("Siyam", 23)
print(person.name)
```

```python
# all the persons variables are publicly available 
dir(person)
```

+ private access modifier 
  + marker: __ (double underscore before variable name)
  + normally we cannot access it outside the class
  + not even derived class

```python
class Person:
    def __init__(self, name, age, cgpa):
        self.__name = name # private variable 
        self.__age = age # private variable 

        self.cgpa = cgpa # public variable
```

```python
person = Person("Siyam", 23, 3.95)

# public variable cgpa present in dir
# but private variable name and age found in dir but in private format
dir(person)
```

+ protected access modifier 
  + marker: _ (single underscore before variable name)
  + normally we cannot access it outside the class
  + but can use derived class

```python
class Person:
    def __init__(self, name, age, cgpa):
        self._name = name # protected variable 
        self._age = age # protected variable 

        self.cgpa = cgpa # public variable

# derived class
class Employee(Person):
    def __init__(self, name, age, cgpa):
        super().__init__(name, age, cgpa)

employee = Employee("siyam", 23, 3.95)
print(employee._name)
```

+ Encapsulation with getter and setter

```python
class Person:
    def __init__(self, name, age):
        self.__name = name # private variable 
        self.__age = age # private variable 
    
    # getter method for name: get the value without exposing private variable
    def get_name(self):
        return self.__name
    
    # setter method for name: set the value without exposing private variable
    def set_name(self, name):
        self.__name = name

    # getter method for age
    def get_age(self):
        return self.__age
    
    # setter method for age
    def set_age(self,age):
        if(age>0):
            self.__age = age
        else:
            print("age cannot be negative")
```

```python
person = Person('Siyam',23)
# access and modify private variables using getter and setter
print(person.get_name())
print(person.get_age())

person.set_name('siyam hossain')
print(person.get_name())

person.set_age(21)
print(person.get_age())
person.set_age(-5)
```
