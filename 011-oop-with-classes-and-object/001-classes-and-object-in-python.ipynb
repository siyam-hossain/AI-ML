+ instance variable (attribute of a class) and methods
+ **constructor**: 
  + A constructor is a special type of method used in object-oriented programming languages to initialize objects. The constructor is called automatically every time when an object is created, allowing the object to set initial values for its attributes or perform other setup tasks.
+ **self**
  + In Python, self is used as the first parameter in instance methods to refer to the current object. It allows methods within the class to access and modify the object's attributes, making each object independent of others.

```python
class Dog:
    # constructor: create and initialize a new object from a class
    def __init__(self, name, age):
        # instance variable: self.variable_name
        self.name = name
        self.age = age
```

+ creating objects

```python
dog1 = Dog("Tommy",5)
print(dog1)
# accessing instance variable
print(dog1.name)
print(dog1.age)
```

```python
dog2 = Dog("Don",3)
print(dog2)
print(dog2.name)
print(dog2.age)
```

+ define a class with instance methods

```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # instance methods
    # here we use self keyword in order to access the instance variable
    def bark(self):
        print(f"{self.name} says woof")
```

```python
dog1 = Dog("Tommy",4)
dog1.bark()
```

```python
dog2 = Dog("Thunder",3)
dog2.bark()
```

+ modeling a bank account

```python
# define a class for a bank account
class BankAccount:
    # constructor
    # balance has default value
    def __init__(self,owner, balance=0):
        # instance variable
        self.owner = owner
        self.balance = balance

    # instance methods
    def deposit(self, amount):
        self.balance += amount
        print(f"{amount} is deposited.\nNew balance is: {self.balance}")

    def withdraw(self, amount):
        if (amount<=self.balance and amount>=0):
            self.balance -= amount
            print(f"Withdraw amount: {amount}.\nNew balance: {self.balance}")
        else:
            print("Insufficient balance!")
    
    def get_balance(self):
        return self.balance
```

```python
# crating object
acc1 = BankAccount("siyam",2500)
print(acc1.get_balance())
```

```python
# calling instance methods
acc1.deposit(100)
```

```python
acc1.withdraw(300)
```

```python
acc1.withdraw(2300)
```

```python
acc1.withdraw(5)
```
