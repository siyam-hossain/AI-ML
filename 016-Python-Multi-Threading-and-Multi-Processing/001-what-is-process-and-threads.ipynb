#### **Multi threading and multi processing**

1. Program
2. Process
3. Threads
4. Examples of process and thread

---
**Program:**
+ a program is a sequence of instructions written in programming language.

**Process:**
+ a process is a simply an instance of a program that is being executed.
+ it requires resources like:
  + code segment
  + data segment
  + heap
  + stack
  + register
+ each process requires separate memory space
  + one process cannot corrupt another process
+ i/o requirements 
  + increased execution time to switch between process

**Thread**
+ a thread is a unit of execution with a process
  + single thread process:
    + it has own stack, registers segment
    + where as it shares code segment, data segment, and heap
  + multi-thread:
    + it has multiple thread open
    + each thread has their own stack, register
    + cs, ds, heap are shared
