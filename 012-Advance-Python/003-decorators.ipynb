#### **Decorators**

Decorators are a powerful and flexible feature in python that allows you to modify the behavior of a function or class method. They are commonly used to add functionality to functions or methods without modifying their actual code. This lesson covers the basics of decorators, including how to create and use them.

+ Decorator

```python
def main_welcome(func):

    def sub_welcome_method():
        print("Welcome to the advance python section")
        print("Please learn this concepts properly")
        
        func()
    
    return sub_welcome_method()
```

```python
def course_introduction():
    print("this is an advance python section")

course_introduction()
```

```python
main_welcome(course_introduction)
```

+ below process showcase how we can create decorators

```python
# this is a decorator
@main_welcome
# below function use as a parameter, directly to the main_welcome method
def course_introduction():
    print("This is an advance python section.\nNow we are learning decorators.")
```

```python
@main_welcome
def add():
    print(1+5)
```

```python
def main_welcome(func):

    def sub_welcome_method(a,b):
        print("Welcome to the advance python section")
        print("Please learn this concepts properly")
        
        func(a,b)   
    return sub_welcome_method
```

```python
@main_welcome
def add(a,b):
    print(a+b)

add(5,9)
```

+ best approaches

```python
# decorator
def my_decorator(func):
    def wrapper():
        print("Before the function is called")
        func()
        print("After the function is called")
    return wrapper
```

```python
@my_decorator
def hello():
    print("hello!")
```

```python
hello()
```

+ decorators with arguments

```python
def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)

        # for decorator we return this
        return wrapper
    # for repeat we return this
    return decorator
```

```python
@repeat(3)
def hello():
    print("hello")
```

```python
hello()
```
