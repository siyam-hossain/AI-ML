#### **Generators**

Generators are a simpler way to create iterators. They use the ```yield``` keyword to produce a series of values lazily, which means they generate values on the fly and do not store them in memory

````python
def square(n):
    for i in range(3):
        return i**2
````

````python
square(3)
````

````python
def square(n):
    for i in range(n):
        # create local variable and save the values also, not in memory
        yield i**2
````

````python
square(3)
````

````python
for i in square(3):
    print(i, end=" ")
````

````python
a = square(3)
a
````

````python
next(a)
````

````python
def my_generator():
    # can return multiple value also
    yield 1
    yield 2
    yield 3
    yield 10
````

````python
gen = my_generator()
gen
````

````python
next(gen)
````

````python
for val in my_generator():
    print(val)
````

+ practical: reading large files
  + generators are particularly useful for reading large files because they allow you to process one lne at a time without loading the entire file into memory.

````python
# reading large files

def read_large_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line
````

````python
file_path = '002-large-file.txt'

for line in read_large_file(file_path):
    print(line.strip())
````
