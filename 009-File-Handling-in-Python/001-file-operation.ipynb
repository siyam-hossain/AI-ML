#### **File operation: Read and write files**

File handling is a crucial part of any programming language. Python provides built-in functions and methods to read from and write to files, both text and binary. This lesson will cover the basics of file handling, including reding and writing text files and binary files.

```python
# read a whole file
with open('001-example.txt','r') as file:
    content = file.read()

    print(content)
```

+ read file line by line

```python
with open('001-example.txt','r') as file:
    for line in file:
        print(line.strip()) #.strip() : remove the new line character 
```

+ writing a file (overwriting)

```python
with open('001-example.txt', 'w') as file:
    file.write('Hello world\n')
    file.write('Welcome to file operation\n')
```

+ write a file without overwriting

```python
with open('001-example.txt','a') as file:
    file.write('Append a line in file\n')
```

+ writing list of lines to a file

```python
lines = [
    "first line\n",
    "second line\n",
    "third line\n",
]

with open('001-example.txt','a') as file:
    file.writelines(lines)
```

**Binary files**

+ writing to a binary file

```python
data = b'\x00\x01\x02\x03\x04'

with open('001-example.bin', 'wb') as file:
    file.write(data)
```

+ read a binary file

```python
with open('001-example.bin','rb') as file:
    content = file.read()
    print(content)
```

+ read the content from a source text file and write to ta destination text file
+ copying a text file

```python
with open('001-example.txt','r') as source_file:
    content = source_file.read()

with open('001-destination.txt','w') as destination_file:
    destination_file.write(content)
```

+ Read a text file and count the number of lines, words, and characters.
+ counting lins, words, and characters in a text file

```python
def count_text_file(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        line_count = len(lines)

        word_count = sum(
            len(line.split()) for line in lines
        )
        char_count = sum(
            len(line) for line in lines
        )
    return line_count, word_count, char_count

file_path = '001-example.txt'
lines, words, chars = count_text_file(file_path)

print(f"lines: {lines}, words: {words}, chars: {chars}")
```
