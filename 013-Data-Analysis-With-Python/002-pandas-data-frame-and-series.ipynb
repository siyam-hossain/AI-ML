#### **Pandas DataFrame and Series**

Pandas is a powerful data manipulation library in python, widely used for data analysis and data cleaning. It provides two primary data structures: Series and DataFrame. A series is a one-dimensional array-like object, while a data frame is a two dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (row and columns).

````python
import pandas as pd
````

---
**Series**
  + A pandas series is a one dimensional array-like object that can hold any data type. It is similar to a column in table.

````python
data = list(range(1,6))
print(f"data: {data}")

series = pd.Series(data)
print(f"series:\n{series}")
print(f"type of series: {type(series)}")
````

---
+ series from dictionary elements

````python
data = {
    'a' : 1, 
    'b' : 2,
    'c' : 3,
    'd' : 10
}
dict_series = pd.Series(data)
print(f"series:\n{dict_series}")
print(f"type of series: {type(dict_series)}")
````

````python
data = list(range(10,40,10))
index = list(chr(ch) for ch in range(ord('a'),ord('d')))

print(f"index: {index}, data: {data}")

series = pd.Series(data, index=index)
print(series)
````

---
**DataFrame**

+ create a data frame from a dictionary of list

````python
data = {
    "Name" : ['siyam','hossain', 'alice'],
    'age' : [21, 23, 25],
    'city' : ['Dhaka', 'Sylhet', 'Comilla']
}
df = pd.DataFrame(data)
print(df)
print("--"*20)
print(type(df))
````

````python
import numpy as np
np.array(df)
````
