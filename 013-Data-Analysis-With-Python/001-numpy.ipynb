#### **Numpy**

Numpy is a fundamental library for scientific computing in python. It provides support for arrays and matrices, along with a collection of mathematical functions to operate on these data structures. In this lesson, we will cover the basics of numpy, focusing on arrays and vectorized operations.

**create array using numpy**

+ 1D array

````python
arr1 = np.array([1,2,3,4,5])
print(arr1)
````

+ reshape to 2 dim array

````python
arr2 = np.array([1,2,3,4,5])
arr2 = arr2.reshape(1,5) # 1 row 5 col
print(arr2)
````

````python
print(arr2.ndim)
````

+ one thing remember while reshaping the element must equally distribute in every row

````python
arr2 = np.array(list(range(10)))
print(arr2)

arr2 = arr2.reshape(2,5) # 2 row 5 col
print(arr2)
````

````python
print(f"dimension: {arr2.ndim}")
````

````python
# creating 2d array using manually 
arr2 = np.array(
    [
        list(range(5)),
        list(range(5,10))
    ]
)
print(arr2)
````

````python
print(arr2.ndim)
````

+ ```arange(start, stop, step)```

````python
arr3 = np.arange(0,10,2)
print(arr3)
````

````python
arr3 = arr3.reshape(5,1)
print(arr3)
print(f"dimension: {arr3.ndim}")
````

---
+ attributes of numpy

````python
arr = np.array(
    [
        np.arange(5),
        np.arange(5,10)
    ]
)

print(f"array: {arr}")
print("--"*15)
print(f"shape: {arr.shape}")
print(f"dimension: {arr.ndim}")
print(f"size: {arr.size}")
print(f"data type: {arr.dtype}")
print(f"item size (in bytes): {arr.itemsize}")

````

---
+ numpy vectorized operation

````python
arr1 = np.arange(1,6)
arr2 = np.arange(10,51,10)

print(arr1)
print(arr2)
print("--"*20)

# element wise operation
print(f"addition: {arr1 + arr2}")
print(f"subtraction: {arr1 - arr2}")
print(f"multiplication: {arr1 * arr2}")
print(f"division: {arr1 / arr2}")
print(f"modulus: {arr1 % arr2}")
````

---
+ universal function

````python
arr = np.arange(2,7)

print(arr)
print(f"sqrt: {np.sqrt(arr)}")
print(f"exponential: {np.exp(arr)}")
print(f"sin: {np.sin(arr)}")
print(f"natural log: {np.log(arr)}")
print(f"log10: {np.log10(arr)}")
````

---
+ array slicing and indexing

````python
arr = np.array(
    [
        np.arange(1,5),
        np.arange(5,9),
        np.arange(9,13)
    ]
)
print(f"array: \n{arr}")
````

````python
arr[0]
````

````python
print(arr[0][0])
````

+ with the help of array slicing

````python
# 7 8
# 11 12

print(arr[1:, 2:])
````

````python
# want to extract 
# 3 4
# 7 8
print(arr)
````

````python
print(arr[0:2, 2:])
````

````python
# diagonal element only
row = arr.shape[0]
col = arr.shape[1]

print(f"row: {row}")
print(f"col: {col}")

print("diagonal elements: ", end="")
for i in range(row):
    print(arr[i,i], end=" ")
````

````python
print(arr)
# extract 
# 2 3
# 6 7
# 10 11
print("--"*20)
print(arr[0:, 1:3])
````

---
+ **statistical concepts** 
  + Normalization : $$Z = \frac{X - \mu}{\sigma}$$


````python
data = np.arange(1,6)

mean = np.mean(data)
std_dev = np.std(data)

normalized_data = (data - mean) / std_dev
print("Normalized data: ", normalized_data)
````

````python
data = np.arange(1, 11)

print(f"data: {data}")

mean = np.mean(data)
median = np.median(data)
std_dev = np.std(data)
variance = np.var(data)

print(f"mean: {mean}")
print(f"median: {median}")
print(f"standard deviation: {std_dev}")
print(f"variance: {variance}")
````

+ logical operation

````python
data = np.arange(1, 11)

print(data>5)
print(data[data>5])

print(data[(data>=2) & (data%2==0)])
````
