#### **Data visualization with seaborn**

Seaborn is a python visualization library based on matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code. In this lesson, we will cover the basics of seaborn, including creating various types of plots and customizing them.

```python
import seaborn as sns
```

---
+ basic plot

```python
tips = sns.load_dataset('tips')
tips
```

+ scatter plot
  + Seaborn is built on top of matplotlib, so figure-level controls like title, show, labels, etc. come from plt, not sns.

```python
import matplotlib.pyplot as plt
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("Scatter plot of total bills vs tip")
plt.show()
```

---
+ line plot

```python
sns.lineplot(x='size', y='total_bill',data=tips)
plt.title("Line plot of total bill by size")
plt.show()
```

---
**Categorical plot**

+ bar plot

```python
sns.barplot(x='day',y='total_bill',data=tips,color='red')
plt.title("Bar plot of total bill by day")
plt.show()
```

+ box plot

```python
sns.boxplot(x='day', y='total_bill', data=tips,color='red')
plt.show()
```

+ violin plot

```python
sns.violinplot(x='day', y='total_bill', data=tips,color='orange')
plt.show()
```

---
+ histogram

```python
sns.histplot(tips['total_bill'], bins=10, kde=True)
plt.show()
```

```python
sns.histplot(tips['total_bill'], bins=10, kde=False, color='#EF8182')
plt.show()
```

---
+ kde plot

```python
tips.head()
```

```python
sns.kdeplot(tips['total_bill'],fill=True,color='red')
plt.show()
```

---
+ pair plot

```python
sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips, plot_kws={'color':'red'})
plt.show()
```

---
+ heat map for correlation

```python
corr = tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```

```python
import pandas as pd
sales_df = pd.read_csv('004-sales-data.csv')
sales_df.head()
```

+ plot total sales by product
  + estimator uses aggregator function

```python
plt.figure(figsize=(10,6))

sns.barplot(x='Product Category', y='Total Revenue', data=sales_df, estimator=sum, color='orange')

plt.title('Total sales by product')
plt.xlabel('Product')
plt.ylabel('Total sales')
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.barplot(x='Region', y='Total Revenue', data=sales_df, estimator=sum, color='#05C274')

plt.title('Total sales by Region')
plt.xlabel('Region')
plt.ylabel('Total sales')
plt.show()
```
